# 7. Train and Test

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=15)
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(x_train,y_train)
y_pred = clf.predict(x_test)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score

#Normalize the data attributes for the Iris dataset.
from sklearn.datasets import load_iris
from sklearn import preprocessing
# load the iris dataset
iris = load_iris()
print(iris.data.shape)
# separate the data from the target attributes
x = iris.data
y = iris.target
# normalize the data attributes
normalized_x = preprocessing.normalize(x)

```

```python
print('Accuracy Score : ' + str(accuracy_score(y_test, y_pred)))
print('Precision Score : ' + str(precision_score(y_test, y_pred, pos_label='Presence')))
print('Recall Score : ' + str(recall_score(y_test, y_pred, pos_label='Presence')))
print('F1 Score : ' + str(f1_score(y_test, y_pred, pos_label='Presence')))
from sklearn.metrics import confusion_matrix
print('Confusion Matrix : \n' + str(confusion_matrix(y_test,y_pred)))
     
```

```python
from sklearn.model_selection import GridSearchCV
# Define the model
clf = LogisticRegression()

# Define the parameter grid
grid_values = {'penalty': ['l1', 'l2'], 'C': [0.001, 0.009, 0.01, 0.09, 1, 5, 10, 25, 15, 100]}

# Perform grid search with cross-validation
grid_clf_acc = GridSearchCV(clf, param_grid=grid_values, scoring='accuracy')

# Fit the model
grid_clf_acc.fit(x_train, y_train)

# Predict values based on the best found parameters
y_pred = grid_clf_acc.predict(x_test)
```
